## GAIA!

_Gaia_ (cuyo nombre era originalmente una sigla de [_Global Astrometric Interferometer for Astrophysics_](http://www.esa.int/Our_Activities/Space_Science/Gaia_overview)) es una misión de la ESA lanzada el 19 de diciembre de 2013, cuyo termino se estima el [2018](http://sci.esa.int/gaia/47354-fact-sheet/). Su [objetivo principal](http://www.esa.int/Our_Activities/Space_Science/Gaia/Science_objectives) es inspeccionar unas cien mil millones de estrellas de la Vía Láctea (además de galaxías cercanas a la nuestra), para hacer un mapa 3D preciso y responder preguntas de nuestro origen y evolución. Para ello Gaia determina la posición, el movimiento, el brillo y el color.

Además Gaia tiene otros [objetivos](http://sci.esa.int/gaia/28890-objectives/):
- Realizar una historia de la Vía Láctea.
- Detectar planetas del tamaño de Júpiter a una distancia de diez millones unidades astronómicas aproximadamente con un período orbital entre 1.5 y 9 años terrestres. Se estima que se detectarán unos 10000 planetas.
- Detectar enanas marrones paseando por el espacio. Además de otras orbitando estrellas.
- Detectar asteroides.
- Detectar supernovas.
- Testear la relatividad general.

El catálogo disponible es el del [GDR1](https://gaia.aip.de/cms/data/gdr1/), basado enell TGAS (_Tycho-Gaia Astrometric Solution_), el cual contiene estrellas en común con Tycho-2 e Hipparcos.

Por otro lado, TGAS tiene más de 200000 estrellas en común con el catálogo [RAVE](https://ui.adsabs.harvard.edu/#abs/2016arXiv160902914C/abstract) (_RAdial Velocity Experiment_), que contiene datos espectroscópicos (temperatura efectiva, gravedad superficial y abundancia química). En conjunto nos permite realizar un análisis químico-dinámico de la Vía Láctea.

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
fig.subplots_adjust(hspace=.5)
axs = [ax1, ax2]
axs_hess = [ax3, ax4]

# Obtenemos la temperatura y la magnitud.
teff = raveon_cat[m1]['TEFF']
mod = np.log10(1 / tgas_cat[m2]['parallax']) - 1
mg = tgas_cat[m2]['phot_g_mean_mag'] - 5 * mod

# Obtenemos los valores para el diagrama de Hess, eliminando nan.
delnan = np.isnan(teff) + np.isnan(mg)
x = teff[~delnan]
y = mg[~delnan]

# Estimamos el histograma 2D
nbins = 100
H, xedges, yedges = np.histogram2d(x, y, bins=nbins)
 
# H needs to be rotated and flipped
H = np.rot90(H)
H = np.flipud(H)
 
# Mask zeros
Hmasked = np.ma.masked_where(H == 0, H) # Mask pixels with a value of zero


# El otro diagrama
for ind in range(2):
    axs[ind].plot(teff, mg, '.', color=main_color, alpha=0.2)
    if ind != 1:
        axs[ind].invert_yaxis()
    else:
        axs[ind].set_ylim((25, 5))
        axs[ind].set_xlim((3000, 8000))

# Diagramas de Hess
for ind in range(2):
    # Plot 2D histogram using pcolor
    im = axs_hess[ind].pcolormesh(xedges, yedges, Hmasked)
    fig.colorbar(im, ax=axs_hess[ind])
    if ind != 1:
        axs_hess[ind].invert_yaxis()
    else:
        axs_hess[ind].set_ylim((25, 5))
        axs_hess[ind].set_xlim((3000, 8000))

plt.show()
```

Utilizando las transformaciones utilizadas en Bovy et Al [2009](https://arxiv.org/pdf/0905.2980.pdf)

```python
from astropy import units as u
from astropy.coordinates import SkyCoord

def obs_to_galactic(v_r, mu_ra, mu_dec, ra, dec, plx):
    k = 4.74047  # 1 ua/yr en km/s 
    # Centro de la galaxia
    center_J200 = SkyCoord('12h51m26.282s', '+27d07m42.01s', frame='icrs')
    cen_ra = center_J200.ra.radian
    cen_dec = center_J200.dec.radian
    zero_l = np.radians(122.932)
    
    costh = np.cos(zero_l)
    sinth = np.sin(zero_l)
    
    cosdec = np.cos(cen_dec)
    sindec = np.cos(cen_dec)
    
    cosra = np.cos(cen_ra)
    sinra = np.cos(cen_ra)
    
    # Matrices T
    t1 = np.matrix([[costh, sinth, 0],
                    [sinth, -costh, 0],
                    [0, 0, 1]])
    t2 = np.matrix([[-sindec, 0, cosdec],
                    [0, 1, 0],
                    [cosdec, 0, sindec]])
    t3 = np.matrix([[cosra, sinra, 0],
                    [-sinra, cosra, 0],
                    [0, 0, 1]]) 
    
    # Matrices A
    csdec = np.cos(dec)
    sndec = np.cos(dec)
    
    csra = np.cos(ra)
    snra = np.cos(ra)

    a1 = np.matrix([[csra, -snra, 0],
                    [snra, csra, 0],
                    [0, 0, 1]])
    a2 = np.matrix([[csdec, 0, -sndec],
                    [0, 1, 0],
                    [sndec, 0, csdec]])
    
    t = t1 * t2 *t3
    a = a1 * a2
    
    vect = np.array([v_r,
                     (k / plx) * mu_ra * csdec,
                     (k / plx) * mu_dec])

    return np.dot(t * a, vect)
```

```python
ra = raveon_cat[m1]['RA']
dec = raveon_cat[m1]['DEC']
v_r = raveon_cat[m1]['HRV']

plx = tgas_cat[m2]['parallax']
mu_ra = tgas_cat[m2]['pmra']
mu_dec = tgas_cat[m2]['pmdec']

vels = list()


for ind in range(0, len(ra)):
    v_xyz = obs_to_galactic(v_r[ind], mu_ra[ind],
                            mu_dec[ind], ra[ind],
                            dec[ind], plx[ind])
    vels.append(v_xyz[0: 1])
print("Hola")
```
