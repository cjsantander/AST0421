## Seguimiento espectroscópico de candidatos a binarias eclipsantes.

Con las observaciones hechas en la parte 1, en particular de Bravo ([2016](https://github.com/MBravoS/T4.1/blob/master/T4_1.ipynb)) tomamos los espectros de las binarias eclipsantes, con el telescopio ESO 50cm, en el siguiente orden:

1. HD198044
2. HD205403
3. HD222891

Sin embargo para la primera imagen, el espectro se veía muy débil. Pensamos que podría deberse a la iluminación de la habitación en donde se encontraba el espectógrafo, pero al corregir ese error las imagenes seguían siendo tenues.

Cada vez que se realizaba una imagen de una nueva estrella binaria primero tomábamos una imagen de la lámpara, después una imagen Dark y finalmente el espectro de la binaria.

Con los datosde pucheros y coralie se obtuvieron las siguientes velocidades:

```python
from astropy.time import Time, TimeDelta
from astropy.io import ascii
from scipy.optimize import least_squares
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


# Name: (MJD, period [d])
ephe = {'HD198044': (54450.509489, 5.15195),
        'HD205403': (54123.634818, 2.4449),
        'HD222891': (54150.650434, 1.59495)}
day = 86400  # Seconds

# pucheros.dat
# coralie.dat
# Name: [(MBJD, Velocity)]
ephe_data  = {'HD198044': [],
              'HD205403': [],
              'HD222891': []}

# Extraemos los datos de pucheros y coralie.
for file in ["pucheros.dat", "coralie.dat"]:
    data = ascii.read(file)
    for row in data:
        name = row['CANDIDATE']
        if name in ephe.keys():
            ephe_data[name].append((row['MBJD'], row['RV']))

# Buscamos la curva que se ajuste a los valores
sin_fit = {}
def func_sin(x, y, period):
    omega = 2 * np.pi / period
    return y[:,1] - x[0] * np.sin(omega * y[:,0] + x[1]) 

for name in ephe.keys():
    x0 = np.array([50, 2])
    y = np.array(ephe_data[name])
    res_1 = least_squares(func_sin, x0, args=(y,ephe[name][1]))
    sin_fit[name] = res_1.x


    
for name in ephe.keys():
    values = np.array(ephe_data[name])
    var_min = np.min(values[:,0])
    var_max = np.max(values[:,0])
    t = np.linspace(var_min, var_max, 1000)
    omega = 2 * np.pi / ephe[name][1]
    
    
    h = 5
    w = 20
    fig, ax = plt.subplots(figsize=(w, h))
    

    ax.set_xlabel('MBJD')
    ax.set_ylabel('Velocidad radial km/s')
    ax.set_title("Velocidad radial de {}".format(name))
    ax.plot(t, sin_fit[name][0] * np.sin(sin_fit[name][1] + t * omega))
    ax.scatter(values[:,0], values[:,1])

    plt.xticks(rotation=90)
    max_xticks = 50
    xloc = plt.MaxNLocator(max_xticks)
    ax.xaxis.set_major_locator(xloc)
    plt.show()
```
